```python
import pandas as pd
import datetime
import numpy as np
```

```python
doorA_lambda = 50
doorB_lambda = 70
doorC_lambda = 28
doorD_lambda = 100
doorE_lambda = 42

num_time = 31*24*2
A = np.random.poisson(doorA_lambda, num_time)
B = np.random.poisson(doorB_lambda, num_time)
C = np.random.poisson(doorC_lambda, num_time)
D = np.random.poisson(doorD_lambda, num_time)
E = np.random.poisson(doorE_lambda, num_time)
```

```python
dt = datetime.datetime(2017, 07, 01)
end = datetime.datetime(2017, 07, 31, 23, 59, 59)
step = datetime.timedelta(minutes = 30)

minHour = 8
maxHour = 21
sigma = 1
mu = 14.5
result = []
i = 0
while dt < end:
    result.append(dt.strftime('%Y-%m-%d %H:%M'))
    x = dt.hour
    p = np.exp(-np.power((x-mu),2)/(2*np.power(sigma,2)))/np.sqrt(2*np.pi*np.power(sigma,2))+1
    if dt.hour<minHour or dt.hour>=maxHour:
        A[i]=B[i]=C[i]=D[i]=E[i]=0
    if dt.weekday()>3: #Weekend
        A[i]=int(A[i]*1.2)
        B[i]=int(B[i]*1.2)
        C[i]=int(C[i]*1.2)
        D[i]=int(D[i]*1.2)
        E[i]=int(E[i]*1.2)
#     if dt.hour<11 or dt.hour>19: #Very early or very late
    A[i]=int(A[i]*p)
    B[i]=int(B[i]*p)
    C[i]=int(C[i]*p)
    D[i]=int(D[i]*p)
    E[i]=int(E[i]*p)        
    i += 1
    dt += step
```

```python
mall_data = pd.concat([pd.DataFrame(result),pd.DataFrame(A),pd.DataFrame(B),pd.DataFrame(C), pd.DataFrame(D), pd.DataFrame(E)], axis = 1)
```

```python
mall_data.columns = ['Date', 'DoorA', 'DoorB', 'DoorC', 'DoorD', 'DoorE']
mall_data
```

```python
melted = pd.melt(mall_data, id_vars=['Date'], var_name='Door', value_name='Count')
```

```python
df = melted.sort_values(by = ['Date', 'Door']).reset_index(drop = True)
```

```python
df.to_csv("door_data.csv")
```
